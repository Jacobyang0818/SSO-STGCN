```python
import pandas as pd
import re

# 讀取文件內容
file_path = "paste.txt"
with open(file_path, "r", encoding="utf-8") as file:
    log_data = file.readlines()

# 初始化變數
data = []
split, feature = None, None

# 解析數據
split_pattern = re.compile(r"Split : (\d+)/ Feature: (\w+)")
log_pattern = re.compile(
    r"Run :\s+(\d+) \| Train_cost :\s+([\d.]+)s \| Train_time :\s+([\d.]+)s \| "
    r"Test_time :\s+([\d.]+)s \| Val_acc_4c :\s+([\d.]+) \| Val_acc_3c :\s+([\d.]+) \| "
    r"Val_acc_2c :\s+([\d.]+) \| Test_acc_4c :\s+([\d.]+) \| Test_acc_3c :\s+([\d.]+) \| "
    r"Test_acc_2c :\s+([\d.]+) \| F1 :\s+([\d.]+) \| F1_4c :\s+([\d.]+) \| F1_3c :\s+([\d.]+) \| Precision :\s+([\d.]+) \| "
    r"Recall :\s+([\d.]+) \| Log : ([\w/_.-]+)"
)

for line in log_data:
    # 嘗試匹配 Split 和 Feature
    split_match = split_pattern.search(line)
    if split_match:
        split, feature = split_match.groups()
        continue

    # 嘗試匹配 Run 記錄
    log_match = log_pattern.search(line)
    if log_match:
        (run, train_cost, train_time, test_time, val_acc_4c, val_acc_3c, val_acc_2c,
        test_acc_4c, test_acc_3c, test_acc_2c, f1, f1_4c, f1_3c, precision, recall, log_file) = log_match.groups()
        
        data.append([
            split, feature, int(run), float(train_cost), float(train_time), float(test_time), 
            float(val_acc_4c), float(val_acc_3c), float(val_acc_2c), 
            float(test_acc_4c), float(test_acc_3c), float(test_acc_2c), 
            float(f1), float(f1_4c), float(f1_3c), float(precision), float(recall), log_file
        ])

# 創建 DataFrame
columns = ["Split", "Feature", "Run", "Train Cost (s)", "Train Time (s)", "Test Time (s)", 
           "Val Acc 4C", "Val Acc 3C", "Val Acc 2C", 
           "Test Acc 4C", "Test Acc 3C", "Test Acc 2C", 
           "F1 Score", "F1_4C", "F1_3C", "Precision", "Recall", "Log File"]
df = pd.DataFrame(data, columns=columns)

csv_path = "experiment_results.csv"
df.to_csv(csv_path, index=False, encoding="utf-8-sig")
print("Done")
```
